### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

try:
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    colors = ['r', 'g', 'b', 'y']

    plt.bar(categories, values, color=colors)
    plt.title('Bar Plot of Categories')
    plt.show()
except Exception as e:
    print(f"Error: {e}")
```
